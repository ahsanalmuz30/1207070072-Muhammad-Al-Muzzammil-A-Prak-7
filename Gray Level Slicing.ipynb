```python
import cv2 #mengimpor gambar
import numpy as np # data dalam bentuk array
from skimage import data #menggunakan modul data yang disediakan oleh pustaka skimage
import matplotlib.pyplot as plt #grafik
%matplotlib inline
```

```python
img = cv2.imread ("ahsan.jpg", cv2.IMREAD_GRAYSCALE) #mengubah gambar menjadi abu abu

row, column = img.shape #untuk mendapatkan dimensi (jumlah baris dan kolom) dari suatu citra atau gambar yang direpresentasikan oleh variabel img

img1 = np.zeros((row,column),dtype = 'uint8') # untuk membuat array Numpy dua dimensi yang berukuran row x column dengan tipe data uint8 
 

min_range = 10
max_range = 60
 

for i in range(row): #Melakukan iterasi pada setiap pixel gambar 
    for j in range(column):
        if img[i,j]>min_range and img[i,j]<max_range: #untuk melakukan pengujian kondisi pada nilai piksel tertentu dalam gambar yang direpresentasikan oleh variabel img
            img1[i,j] = 255
        else:
            img1[i,j] = 0
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12)) #membuat gambar dengan multiple subplot (subgambar) dalam satu tampilan
ax = axes.ravel() #untuk meratakan (flatten) objek array axes menjadi array satu dimensi yang disimpan dalam variabel ax

ax[0].imshow(img, cmap=plt.cm.gray) #untuk menampilkan gambar (image) img pada sumbu (axis) indeks ke-0 dari objek axes. Gambar ditampilkan dalam skala abu-abu (grayscale)
ax[0].set_title("Citra Input") #judul
ax[1].hist(img.ravel(), bins=256) # untuk membuat histogram dari gambar img dan menampilkannya pada sumbu (axis) indeks ke-1 dari objek axes
ax[1].set_title('Histogram Input') #judul

ax[2].imshow(img1, cmap=plt.cm.gray)
ax[2].set_title("Citra Output")
ax[3].hist(img1.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```
